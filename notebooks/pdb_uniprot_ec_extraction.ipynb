```python
enz = 'translocases3'
files_parent_dir = f"data/{enz}/"
```

```python
pdb_zipped_files = [file for file in os.listdir(files_parent_dir) if '.pdb.gz' in file]
```

```python
def extract_uniprot_ec_from_zipped_pdb(zipped_files_list, files_parent_dir):
    pdb_uniprot_ec =[]
    for file_name in zipped_files_list:
        os.system(f"gzip -dk {files_parent_dir}{file_name}")
        pdb_file = file_name.split('.gz')[0]
        # Get pdb id
        pdb_name = file_name.split('.pdb')[0]
        # Read pdb file
        pdb_str = open(f"{files_parent_dir}{pdb_file}", "r").read()
        # Get uniprot id
        unip = re.search('UNP +(\w+)', pdb_str)
        uniprot_id = unip.group(1)
        print(pdb_name)
        # Get Enzyme Classification number
        ec = re.search('EC: +(\d+)[.](\d+)[.](\d+)[.](\d+)', pdb_str)
        if ec is None:
            ec_num=""
        else:
            ec_num = str(ec.group())[4:]
        os.remove(f"{files_parent_dir}{pdb_file}")
        pdb_uniprot_ec.append([pdb_name,uniprot_id, ec_num])
    df_pdb_uniprot_ec = pd.DataFrame(pdb_uniprot_ec,columns=['pdb_id','uniprot_id','ec_num'])
    df_pdb_uniprot_ec.to_csv(f"{files_parent_dir}df_pdb_uniprot_ec.csv",index=False)
```
